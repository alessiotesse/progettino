```python
og1=og[['team','no_games','no_hosted']]  #creo la variabile su quante volte le nazioni hanno gareggiato nelle discipline olimpiche e quante volte le nazioni hanno ospitato un'olimpiade
og1
```

In questo grafico ho rappresentato tutte le nazioni che hanno ospitato, che hanno vinto le medaglie e in quanti sport hanno gareggiato

```python
og1.describe()  #qui descrivo il totale delle nazioni che hanno partecipato alle olimpiadi, la media e la mediana di quante volte hanno partecipato alle discipline sportive e qunate volte hanno ospitato le olimpiadi
```

```python
og.groupby('team').mean()  #qui raccolgo in ordine alfabetico tutte le nazioni che hanno partecipato ai giochi olimpici, hanno vinto le medaglie olimpiche e che le hanno ospitato
```

```python
og11=og1[og1['team'].isin(['Russia','France','Great Britain','United States'])]  #creo la variabile del filtro concentrandomi sulle potenze e qunate volte hanno ospitato un'olimpiade e in quanti sport hanno gareggiato
og11
```

```python
og11.plot.bar(rot=70, figsize=(5,5), fontsize="medium")
```

In questo grafico a barre si può notare come le potenze olimpiche abbiano ospitato e in quante discipline abbiano gareggiato nelle discipline sportive

# Il caso cinese

```python
import matplotlib.pyplot as plt
plt.plot([1952,1984,1988,1992,1996,2000,2004,2008,2012,2016,2020],[0,32,28,54,50,58,64,100,91,70,88])
plt.ylabel('China_medal_overall')
plt.xlabel('olympic_year')
plt.show
```

E' interessante il caso della Cina che nel 1952 fu il primo banco di prova della Cina comunista ma ebbe scarsi risultati ma nonostante ciò la classe dirigente cinese ottenne una vittoria dal punto di vista diplomatico e assistette alle vittorie sportive del blocco dell'Europa dell'est. Questo convinse il governo a dar vita a un sistema sportivo centralizzato ispirato al modello sovietico e lo sport divenne nella propaganda del partito "la continuazione della politica". Ma i giochi del 1952 furono gli ultimi fino al 1984. Con le olimpiadi di Los Angeles nel '84 è coinciso con la strategia della gold medal strategy, cioè un piano di investimenti per vincere più medaglie d'oro degli altri paesi e diventare una potenza olimpica. Questa visione è condivisa anche dal PCC, che ha iniziato ad impiegare lo sport non solo per migliorare l’immagine della Cina e produrre soft power in politica estera, ma anche come strumento di soft power domestico, con l’obiettivo di consolidare la posizione del governo di fronte ai cittadini. Secondo un articolo di Hannah Beech del NYT la Cina ha investito negli sport meno noti che sono sotto finanziati in Occidente. Il tallone d'Achille della Cina olimpica è rappresentato dagli sport di squadra
